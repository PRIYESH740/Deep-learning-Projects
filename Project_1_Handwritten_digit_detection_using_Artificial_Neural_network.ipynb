Choosing the model/ Apply the ANN

#keras sequential units (Neuron)

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),  # Explicit Input layer
    keras.layers.Dense(units=10, activation='sigmoid')  # Dense layer without input_shape
])
model.compile(
    optimizer='Adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
tf.config.experimental_run_functions_eagerly(True)
```

Trained the model

```python
model.fit(flat_x_train,y_train,epochs=5)
```

Feature scaling on dataset

```python
new_x_train=x_train/255
new_x_test=x_test/255
```

```python
# Flattening the x_train & x_test
new_flat_x_train=new_x_train.reshape(len(new_x_train),28*28)
new_flat_x_test=new_x_test.reshape(len(new_x_test),28*28)
```

```python
#Trained the model after feature scaling
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),  # Explicit Input layer
    keras.layers.Dense(units=10, activation='sigmoid')  # Dense layer without input_shape
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(new_flat_x_train,y_train,epochs=5)
```

#checking accuracy

```python
model.evaluate(new_flat_x_test,y_test)
```

Prediction

```python
y_pred=model.predict(new_flat_x_test)
```

```python
#checking the value of predict  'g value'
y_pred[0]
```

```python
#find the index value of max 'g value' 
np.argmax(y_pred[0])
```

```python
#list out the index value of max all 'g value' 
y_pred_label=[np.argmax(i) for i in y_pred]
```

```python
#show all the index
y_pred_label
```

```python
#confusion matrix 
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_label)
```

```python
#plot heatmap to actual value and predict value is same how many times 
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Actual Value')
plt.ylabel('Predicted value')
```

#Multiple hidden layers

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),  # Explicit Input layer
    keras.layers.Dense(units=10, activation='sigmoid'),
    #second hidden layer
    keras.layers.Dense(units=100, activation='sigmoid') 
    
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(new_flat_x_train,y_train,epochs=5)
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),  # Explicit Input layer
    keras.layers.Dense(units=100, activation='sigmoid'),
    #second hidden layer
    keras.layers.Dense(units=100, activation='sigmoid') 
    
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(new_flat_x_train,y_train,epochs=5)
```

```python
#without using reshape to flatten the train,test matrix 
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),  
    keras.layers.Dense(units=100, activation='sigmoid'),
    #second hidden layer
    keras.layers.Dense(units=100, activation='sigmoid') 
    
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=5)
```
